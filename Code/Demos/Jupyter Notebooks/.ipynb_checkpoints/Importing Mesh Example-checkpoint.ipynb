## Importing/Exporting Mesh, Boundary and Subdomain Labels into FEniCS in Parallel

As dolfin-convert is no longer maintained (meshio is recommended replacement, we will use a compatible version of gmsh to convert the .msh file to a xml format. This will require using an older version of gmsh (works with v3.1 but not v4.0). FEniCS has the ability to read and write to these file formats.

Convert.msh file to .xml using FEniCS:

From the terminal, type:

#### dolfin-convert meshfile.msh meshfile.xml

If you are running a docker container, you will have to access FEniCS from the terminal of the docker container. You may gain access to the terminal of your docker container by the following:

Type the following into your terminal.

####  docker image ls

It will output a table where you can see the IMAGE_ID of your fenics image. Copy this for the next command in terminal:

#### docker run -i -t IMAGE_ID /bin/bash

This will run the container and your terminal will switch to the container's terminal (kind of like with conda and virtual environments). From there you can navigate to the appropriate directory and execute the dolfin-convert command. Note, Docker containers might not have access to all of your system by default, so some files/folders may be out of reach - this can be easily addressed by installing your docker container with shared access. Please refer to google to clone a docker image for FEniCS with shared access.

Note: The following part of this notebook demonstates how to read in the mesh, subdomain, and boundary labels from a .xml file. You may then convert the .xml file to .HDMF (.h5). These file formats are ideal for large meshes as they offer the ability to operate in parallel whereas .xml files are read in serially. This plays a major role for large meshes when scalability and performance are desired.

```python
from dolfin import *
import matplotlib.pyplot as plt
```

```python
## Specify file name
filename = "./Mesh/3DDomain"
```

```python
## Read in Mesh contained in Mesh.xml file
mesh = Mesh(filename+".xml")                                                  
subdomains = MeshFunction("size_t", mesh, filename + "_physical_region.xml")    
boundaries = MeshFunction("size_t", mesh, filename + "_facet_region.xml")
```

```python
## Write to HDF5 File so that it can be read in parallel
mesh_file = HDF5File(MPI.comm_world, filename+'.h5', 'w')
mesh_file.write(mesh, '/mesh')

# maybe you have defined a mesh-function (boundaries, domains ec.)
# in the xml_mesh aswell, in this case use the following two lines
mesh_file.write(subdomains, "/subdomains")
mesh_file.write(boundaries, "/boundaries")
```

```python
## We can now read the mesh in parallel
mesh_parallel = Mesh()
hdf = HDF5File(MPI.comm_world, filename+".h5", "r")
hdf.read(mesh_parallel, "/mesh", False)
subdomains_parallel = MeshFunction("size_t", mesh_parallel, 3) # Dim = 3 if domain = 3D (CellFunction)
hdf.read(subdomains_parallel, "/subdomains")
boundaries_parallel = MeshFunction("size_t", mesh_parallel, 2) # Dim = 2 if domain = 3D (FacetFunction)
hdf.read(boundaries_parallel, "/boundaries")
```

```python
plot(mesh_parallel)
plt.rcParams['figure.figsize'] = [15, 15]
```
