```python
import matplotlib.pyplot as plt
import numpy as np
from decimal import Decimal
```

```python
DimensionlessHeight = np.genfromtxt("DimensionlessHeight.csv", delimiter=',')
DimensionlessOverpressure = np.genfromtxt("DimensionlessOverpressure.csv", delimiter=',')
PressureScaledbyOverburden = np.genfromtxt("FractionofVerticalStress.csv", delimiter=',')
```

```python
viscosity_ref = [1.0e10,1.0e11,1.0e12,1.0e13,1.0e14,1.0e15,1.0e16,1.0e17,1.0e18,1.0e19]
```

```python
DH = DimensionlessHeight[200:300]
DP = DimensionlessOverpressure[200:300]
```

```python
for i in range(0,len(viscosity_ref)):
    DH = DimensionlessHeight[(100*i+0):(100*i+100)]
    DP = DimensionlessOverpressure[(100*i+0):(100*i+100)]
    plt.figure(figsize=((10,6)))
    plt.plot(DH[DH<40.0],DP[DH<40.0],'r-',label='%.2E' % Decimal(str(viscosity_ref[i])))
    plt.xlabel(r'$\frac{H}{\delta}$',fontsize=18)
    plt.ylabel('Dimensionless Overpressure',fontsize=14)
    plt.title(r'$\zeta = $'+'%.2E' % Decimal(str(viscosity_ref[i])),fontsize=14)
    plt.savefig('Overpressure'+'%.2E' % Decimal(str(viscosity_ref[i]))+'.png')
```

```python
viscosity_ref = [1.0e10,1.0e11,1.0e12,1.0e13,1.0e14,1.0e15,1.0e16,1.0e17,1.0e18,1.0e19]
Porosity = np.linspace(0.001,0.20,100)
Porosity_scaled = np.empty(len(Porosity)*len(viscosity_ref))

for i in range(0,len(viscosity_ref)):
    Porosity_scaled[(100*i+0):(100*i+100)] = Porosity/viscosity_ref[i]
```

```python
plt.scatter(Porosity_scaled[900:]/1.0e-19,DimensionlessOverpressure[900:])
```
